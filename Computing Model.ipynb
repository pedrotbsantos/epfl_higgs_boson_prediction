### Load all the Data

```python
train_data_0,train_data_1,train_data_2,train_data_3 = load_train_datasets()
original_train_data = load_original_train_data()
sorted_train_data = np.asarray(sorted(original_train_data,key=lambda original_train_data: original_train_data[24]))
```

```python
x0 = train_data_0[:,2:]
x1 = train_data_1[:,2:]
x2 = train_data_2[:,2:]
x3 = train_data_3[:,2:]
y0 = train_data_0[:,1]
y1 = train_data_1[:,1]
y2 = train_data_2[:,1]
y3 = train_data_3[:,1]
index0 = train_data_0[:,0]
index1 = train_data_1[:,0]
index2 = train_data_2[:,0]
index3 = train_data_3[:,0]
degrees = [2, 5]
lambdas = np.logspace(-10,0,100)
gammas = np.logspace(-10,0,100)
w0,w1,w2,w3 = load_ws()
```

### Compile with Rigde Regression
We use Rigde Regression with feature expansion. 

```python
print("----------------Model 1/4 --------------------")
w0, loss0, poly0, deg0 = compute_model_ridge(x0,y0,lambdas, degrees)
print("----------------Model 2/4 --------------------")
w1, loss1, poly1, deg1 = compute_model_ridge(x1,y1,lambdas, degrees)
print("----------------Model 3/4 --------------------")
w2, loss2, poly2, deg2 = compute_model_ridge(x2,y2,lambdas, degrees)
print("----------------Model 4/4 --------------------")
w3, loss3, poly3, deg3 = compute_model_ridge(x3,y3,lambdas, degrees)
```
